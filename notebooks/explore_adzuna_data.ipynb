# Raw Data

```python
import json

response_files = sorted(RAW_BASE.glob("*/*response_*.json"))

rows = []
for fp in response_files:
    data = json.loads(fp.read_text(encoding="utf-8"))
    rows.append({
        "file": str(fp),
        "count": data.get("count"),
        "n_results": len(data.get("results", []) or []),
    })
len(rows)
```

```python
rows
```

```python
def req_path_from_res_path(res_path: str) -> Path:
    p = Path(res_path)
    # response_001.json -> request_001_params.json
    name = p.name.replace("response_", "request_").replace(".json", "_params.json")
    return p.with_name(name)

labeled = []
for r in rows:
    res_path = Path(r["file"])
    req_path = req_path_from_res_path(r["file"])
    req = json.loads(req_path.read_text(encoding="utf-8")) if req_path.exists() else {}
    params = req.get("params", {})
    labeled.append({
        "file": r["file"],
        "count": r["count"],
        "n_results": r["n_results"],
        "what": params.get("what"),
        "where": params.get("where"),
        "page_url": req.get("url"),
        "results_per_page": params.get("results_per_page"),
        "sort_by": params.get("sort_by"),
        "what_exclude": params.get("what_exclude"),
    })

labeled
```

```python
ENTRY_WHATS = {"werkstudent data", "praktikum data", "junior data"}

selected = []
for r in rows:
    res_path = Path(r["file"])
    req_path = req_path_from_res_path(r["file"])
    req = json.loads(req_path.read_text(encoding="utf-8")) if req_path.exists() else {}
    params = req.get("params", {})
    what = params.get("what")
    if what in ENTRY_WHATS:
        selected.append(res_path)

len(selected), [str(p) for p in selected]
```

```python
all_ids = []
total_results = 0

for fp in selected:
    data = json.loads(fp.read_text(encoding="utf-8"))
    results = data.get("results", []) or []
    total_results += len(results)
    for job in results:
        if isinstance(job, dict) and "id" in job:
            all_ids.append(job["id"])

unique_ids = set(all_ids)
dup_count = len(all_ids) - len(unique_ids)
dup_rate = (dup_count / len(all_ids)) if all_ids else 0

{
    "selected_files": len(selected),
    "total_results_rows": total_results,
    "ids_collected": len(all_ids),
    "unique_ids": len(unique_ids),
    "duplicate_ids": dup_count,
    "duplicate_rate": round(dup_rate, 4),
}
```

```python
import json
from collections import Counter

# Build a dict of unique jobs by id (keeps first occurrence)
jobs_by_id = {}
for fp in selected:
    data = json.loads(fp.read_text(encoding="utf-8"))
    for job in (data.get("results", []) or []):
        if isinstance(job, dict) and "id" in job and job["id"] not in jobs_by_id:
            jobs_by_id[job["id"]] = job

len(jobs_by_id)
```

```python
# Count top-level key presence across unique jobs
key_counts = Counter()
for job in jobs_by_id.values():
    key_counts.update(job.keys())

# show counts sorted by frequency
sorted(key_counts.items(), key=lambda x: (-x[1], x[0]))
```

```python
from collections import Counter

# Count dict-valued fields and their subkeys across all unique jobs
dict_field_counts = Counter()
subkey_counts = {}  # dict_field -> Counter of subkeys

for job in jobs_by_id.values():
    for k, v in job.items():
        if isinstance(v, dict):
            dict_field_counts[k] += 1
            if k not in subkey_counts:
                subkey_counts[k] = Counter()
            subkey_counts[k].update(v.keys())

dict_field_counts, {k: sorted(subkey_counts[k].items(), key=lambda x: (-x[1], x[0])) for k in subkey_counts}
```

```python
import pandas as pd

# convert dict-of-jobs -> list-of-jobs
jobs_list = list(jobs_by_id.values())

df = pd.json_normalize(jobs_list)  # no explicit loop
df.shape
```

```python
cols = [
    "id", "created", "title", "description", "redirect_url", "salary_is_predicted", "adref",
    "company.display_name",
    "category.label", "category.tag",
    "location.display_name", "location.area",
    "latitude", "longitude", "contract_time", "contract_type",
]
existing = [c for c in cols if c in df.columns]
df_sample = df[existing].copy()
df_sample.head(10)
```
